```python
import mysql.connector
from faker import Faker
import random
from datetime import datetime, timedelta

# Configuração da conexão com o banco de dados
config = {
    'user': 'lucas',  # Substituir pelo usuário do banco
    'password': '',  # Substituir pela senha do banco
    'host': '127.0.0.1',  # Substituir pelo host do banco
    'port': 3306,  # Substituir pela porta do banco
    'database': 'mev_estetica_automotiva'
}


# Inicializar faker e conectar ao banco
faker = Faker()
cnx = mysql.connector.connect(**config)
cursor = cnx.cursor()

# Função para inserir dados na tabela "usuarios"
def inserir_usuarios():
    # Administrador e funcionários
    usuarios = [
        ('Administrador Geral', 'admin@estetica.com', 'senha123', 'administrador'),
        ('Gerente 1', 'gerente1@estetica.com', 'senha123', 'gerente'),
        ('Supervisor 1', 'supervisor1@estetica.com', 'senha123', 'supervisor'),
        ('Funcionario 1', 'funcionario1@estetica.com', 'senha123', 'funcionario'),
        ('Funcionario 2', 'funcionario2@estetica.com', 'senha123', 'funcionario'),
        ('Funcionario 3', 'funcionario3@estetica.com', 'senha123', 'funcionario')
    ]

    # Inserir clientes
    for i in range(100):
        nome = faker.name()
        email = faker.email()
        senha = 'senha123'
        usuarios.append((nome, email, senha, 'cliente'))

    cursor.executemany(
        "INSERT INTO usuarios (nome, email, senha, tipo) VALUES (%s, %s, %s, %s)",
        usuarios
    )
    cnx.commit()
    print(f"{len(usuarios)} usuários inseridos.")

# Função para inserir dados na tabela "servicos"
def inserir_servicos():
    servicos = [
        ('Lavagem Simples', 'Lavagem externa do veículo', 50.00),
        ('Lavagem Completa', 'Lavagem interna e externa', 120.00),
        ('Polimento', 'Polimento completo da pintura', 250.00),
        ('Higienização', 'Higienização interna e ar-condicionado', 150.00),
        ('Cristalização', 'Cristalização da pintura', 300.00)
    ]

    cursor.executemany(
        "INSERT INTO servicos (nome, descricao, preco_base) VALUES (%s, %s, %s)",
        servicos
    )
    cnx.commit()
    print(f"{len(servicos)} serviços inseridos.")

# Função para inserir dados na tabela "produtos"
def inserir_produtos():
    produtos = [
        ('Cera Automotiva', 'Cera de alta performance', 80.00, 50),
        ('Shampoo Automotivo', 'Shampoo especial para lavagem', 30.00, 100),
        ('Aromatizante', 'Aromatizante para carros', 10.00, 200),
        ('Pano de Microfibra', 'Pano para limpeza e polimento', 20.00, 150),
        ('Escova para Estofados', 'Escova para limpeza de estofados', 40.00, 75)
    ]

    cursor.executemany(
        "INSERT INTO produtos (nome, descricao, preco, estoque) VALUES (%s, %s, %s, %s)",
        produtos
    )
    cnx.commit()
    print(f"{len(produtos)} produtos inseridos.")

# Função para inserir dados na tabela "agendamentos"
def inserir_agendamentos():
    agendamentos = []
    for i in range(200):
        cliente_id = random.randint(7, 106)  # IDs de clientes (começam após os funcionários)
        funcionario_id = random.randint(4, 5, 6)  # IDs de funcionários
        servico_id = random.randint(1, 5)  # IDs de serviços
        data_agendamento = faker.date_time_between(start_date="-1y", end_date="now")
        status = random.choice(['pendente', 'confirmado', 'cancelado', 'concluido'])
        observacoes = faker.text(max_nb_chars=50)
        agendamentos.append((cliente_id, funcionario_id, servico_id, data_agendamento, status, observacoes))

    cursor.executemany(
        "INSERT INTO agendamentos (cliente_id, funcionario_id, servico_id, data_agendamento, status, observacoes) VALUES (%s, %s, %s, %s, %s, %s)",
        agendamentos
    )
    cnx.commit()
    print(f"{len(agendamentos)} agendamentos inseridos.")

# Função para inserir dados na tabela "financeiro"
def inserir_financeiro():
    financeiro = []
    data_base = datetime.now() - timedelta(days=365)
    for i in range(12):
        data_registro = data_base + timedelta(days=i * 30)
        receita = random.uniform(10000, 20000)
        custos_fixos = random.uniform(3000, 5000)
        custos_variaveis = random.uniform(2000, 6000)
        lucro_liquido = receita - (custos_fixos + custos_variaveis)
        dividas = random.uniform(1000, 3000)
        fluxo_caixa = lucro_liquido - dividas
        financeiro.append((data_registro.date(), receita, custos_fixos, custos_variaveis, lucro_liquido, dividas, fluxo_caixa))

    cursor.executemany(
        "INSERT INTO financeiro (data_registro, receita, custos_fixos, custos_variaveis, lucro_liquido, dividas, fluxo_caixa) VALUES (%s, %s, %s, %s, %s, %s, %s)",
        financeiro
    )
    cnx.commit()
    print(f"{len(financeiro)} registros financeiros inseridos.")

# Executar todas as funções de inserção
inserir_usuarios()
inserir_servicos()
inserir_produtos()
inserir_agendamentos()
inserir_financeiro()

# Fechar conexão com o banco de dados
cursor.close()
cnx.close()
```

```python
import random
from datetime import datetime, timedelta

# Função para gerar uma data aleatória nos últimos 365 dias
def random_date():
    days_ago = random.randint(0, 365)
    return datetime.now() - timedelta(days=days_ago)

# Parâmetros
num_agendamentos = 100
cliente_ids = list(range(1, 100))  # IDs de clientes (1 a 100)
funcionario_ids = [2, 3, 4, 5, 6]  # IDs de funcionários (2 a 6)
servico_ids = list(range(1, 37))  # IDs de serviços (1 a 36)
status_options = ['pendente', 'confirmado', 'cancelado', 'concluido']
observacoes = [
    "Cliente prefere o serviço no início da manhã.",
    "Cliente quer combinar o pagamento em dinheiro.",
    "Serviço rápido, cliente satisfeito.",
    "Cliente solicitou cancelamento por imprevisto.",
    "Cliente pediu revisão adicional.",
    "Cancelado devido ao clima.",
    "Cliente recomendou o serviço.",
    "Cliente solicitou higienização extra.",
    None  # Para observações vazias
]

# Gerar agendamentos
agendamentos = []
for _ in range(num_agendamentos):
    cliente_id = random.choice(cliente_ids)
    funcionario_id = random.choice(funcionario_ids)
    servico_id = random.choice(servico_ids)
    data_agendamento = random_date()
    status = random.choice(status_options)
    obs = random.choice(observacoes)

    agendamento = (
        cliente_id,
        funcionario_id,
        servico_id,
        data_agendamento.strftime('%Y-%m-%d %H:%M:%S'),
        status,
        obs
    )
    agendamentos.append(agendamento)

# Gerar script SQL
sql_script = "INSERT INTO agendamentos (cliente_id, funcionario_id, servico_id, data_agendamento, status, observacoes, data_criacao) VALUES\n"
values = []
for agendamento in agendamentos:
    cliente_id, funcionario_id, servico_id, data_agendamento, status, obs = agendamento
    obs_value = f"'{obs}'" if obs else "NULL"
    values.append(f"({cliente_id}, {funcionario_id}, {servico_id}, '{data_agendamento}', '{status}', {obs_value}, CURRENT_TIMESTAMP)")

sql_script += ",\n".join(values) + ";"

# Salvar em arquivo
with open("agendamentos_inserts2.sql", "w", encoding="utf-8") as file:
    file.write(sql_script)

print("Script SQL gerado e salvo em 'agendamentos_inserts.sql'.")
```
